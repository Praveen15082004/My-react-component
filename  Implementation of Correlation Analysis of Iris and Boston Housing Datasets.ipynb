```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris, fetch_california_housing

```

```python
# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target
print(iris_df)
iris_corr = iris_df.corr()
```

```python
print(iris_corr)
```

```python
# Plot heatmap for Iris dataset
plt.figure(figsize=(8, 6))
sns.heatmap(iris_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix for Iris Dataset')
plt.show()
```

```python
# Load the California housing dataset
california = fetch_california_housing(as_frame=True)
california_df = california.frame
# Correlation analysis for California housing dataset
california_corr = california_df.corr()

print(california_df)
print(california_corr)
```

```python
# Plot heatmap for California housing dataset
plt.figure(figsize=(10, 8))
sns.heatmap(california_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix for California Housing Dataset')
plt.show()
```
